```python
#Import Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("XAUUSD_Hourly.csv")
```

```python
#Check for Missing Values
df.isnull().sum()
```

```python
#Remove Duplicates
df = df.drop_duplicates()
```

```python
#Initial inspection of the data
print("Initial Data Info:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Handling missing or invalid values
# Replacing any NaN values with appropriate strategies (e.g., forward fill for price columns)
price_columns = ['Open', 'High', 'Low', 'Close', 'Target_Close']
for col in price_columns:
    df[col] = df[col].fillna(method='ffill') 
df['TickVolume'] = df['TickVolume'].fillna(df['TickVolume'].median())  
df['Spread'] = df['Spread'].fillna(df['Spread'].mode()[0])  
```

```python
#Converting datetime column to proper format
df['datetime'] = pd.to_datetime(df['datetime'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
```

```python
# Verify datetime matches Date and Time columns
df['constructed_datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%Y.%m.%d %H:%M:%S', errors='coerce')
datetime_mismatch = (df['datetime'] != df['constructed_datetime']).sum()
print(f"\nNumber of datetime mismatches: {datetime_mismatch}")
```

```python
# Drop Date, Time, and constructed_datetime as they are redundant
df = df.drop(['Date', 'Time', 'constructed_datetime'], axis=1)
```

```python
#Dropping Volume column (all values are 0)
df = df.drop('Volume', axis=1)
```

```python
#Checking for outliers using IQR method
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]
    return len(outliers)

for col in price_columns:
    outlier_count = detect_outliers(df, col)
    print(f"Outliers in {col}: {outlier_count}")
```

```python
#Handling Spread inconsistencies (values are mostly 5 or 7, with rare 6 or 8)
# Replace rare values (6, 8) with the mode
spread_mode = df['Spread'].mode()[0]
df['Spread'] = df['Spread'].replace([6, 8], spread_mode)
```

```python
#Checking for duplicates
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")
df = df.drop_duplicates()
```

```python
# Final inspection of cleaned data
print("\nCleaned Data Info:")
print(df.info())
print("\nFirst few rows of cleaned data:")
print(df.head())
```

```python
#Saving the cleaned dataset
df.to_csv('XAUUSD_Hourly_Cleaned.csv', index=False)
print("\nCleaned data saved to 'XAUUSD_Hourly_Cleaned.csv'")
```
